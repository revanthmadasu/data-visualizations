```python
# Stacked Bar Plot
# Party Preference by different ethnicity
df = M1_M2_M3_M4_M5  # Creating a DataFrame 'df' by combining data from M1 to M5
x = df['Q3']  # Extracting the 'Q3' column as x
y1 = df['Q19']  # Extracting the 'Q19' column as y1
y2 = df['Q17']  # Extracting the 'Q17' column as y2

def get_party_type(row):  # Defining a function to determine party type preference based on 'Q19' and 'Q17' values
    party_type = ""
    if row['Q19'] > 0 and row['Q17'] > 0:
        party_type = "Both Destination and Themed"
    elif row['Q19'] > 0:
        party_type = "Themed"
    elif row['Q17'] > 0:
        party_type = "Destination"
    else:
        party_type = "Undecided/uninterested"
    return party_type

df['party_type_pref'] = df.apply(get_party_type, axis=1)  # Adding a new column 'party_type_pref' based on the function

grouped_data = df.groupby(['Q3', 'party_type_pref']).size().reset_index(name='count')  # Grouping data by 'Q3' and 'party_type_pref' and counting occurrences

ethnicities = grouped_data['Q3'].unique()  # Extracting unique ethnicities from the grouped data

nodes = []  # Initializing a list to store nodes for each ethnicity
for ethnicity in ethnicities:
    node = {
        'eth': ethnicity,
        'Both Destination and Themed': 0,
        'Themed': 0,
        'Destination': 0,
        'Undecided/uninterested': 0
    }
    ethnicity_data = grouped_data[grouped_data['Q3'] == ethnicity]
    for index, row in ethnicity_data.iterrows():
        node[row['party_type_pref']] = row['count']
    nodes.append(node)

party_theme_by_ethnicity = pd.DataFrame(nodes)  # Creating a DataFrame 'party_theme_by_ethnicity' from the nodes

plt.bar(party_theme_by_ethnicity['eth'], party_theme_by_ethnicity['Undecided/uninterested'])  # Creating a bar for 'Undecided/uninterested'
plt.bar(party_theme_by_ethnicity['eth'], party_theme_by_ethnicity['Destination'], bottom=party_theme_by_ethnicity['Undecided/uninterested'])  # Creating a stacked bar for 'Destination'
plt.bar(party_theme_by_ethnicity['eth'], party_theme_by_ethnicity['Themed'], bottom=party_theme_by_ethnicity['Undecided/uninterested'] + party_theme_by_ethnicity['Destination'])  # Creating a stacked bar for 'Themed'
plt.bar(party_theme_by_ethnicity['eth'], party_theme_by_ethnicity['Both Destination and Themed'], bottom=party_theme_by_ethnicity['Undecided/uninterested'] + party_theme_by_ethnicity['Destination'] + party_theme_by_ethnicity['Themed'])  # Creating a stacked bar for 'Both Destination and Themed'

plt.legend(['Undecided/uninterested', "Destination", "Themed", "Both Destination and Themed"], loc='upper left', fontsize=12)  # Adding a legend to the plot
plt.xlabel('Ethnicities')  # Adding a label for the x-axis
plt.title('Party Preference by Ethnicity')  # Adding a title to the plot
plt.xticks(range(1, 14), ['American Indian', 'African American', 'Mexican American', 'Other Hispanic', 'Asian', 'Southeast Asian', 'Pacific Islander', 'Filipino', 'Multiple Ethnicities', 'White', 'Other/Not Stated', 'International', 'Prefer not to say'], rotation=45)  # Setting custom x-axis ticks and labels
plt.tight_layout()  # Adjusting layout for better display
plt.savefig("7-Party_Preference_Ethnicity-Python.png")  # Saving the plot as an image
plt.show()  # Displaying the plot
```
