```python
import pandas as pd # An "import" statement is used for importing a library for data processing called "pandas" and assigning it to a variable called "pd".
import folium # An "import" statement is used for importing a library for visualizing geospatial data called "folium".
```

```python
# Create a base map centered around San Diego, which is where all of the target customers for the Use Case focus groups were selected from.
m = folium.Map(location=[32.7157, -117.1611], zoom_start=9)

# Use the read_excel() function from a library called pandas to read in a file that contains the coordinates from your team's Use Case M7 dataset that Dr. Mead posted on your team's Bb File Exchange.
coordinates = pd.read_excel("1 - Elite Party Services - M7 - Geographical Data.xlsx") # Change the filename to whatever yours is called. In this example code it is for a random use case.

# Use a for loop to iterate over the rows of the DataFrame and add markers.
for index, row in coordinates.iterrows():
    latitude = row["Latitude"]
    longitude = row["Longitude"]
    folium.Marker(location=[latitude, longitude]).add_to(m)

# Use the save() function to save the map to an HTML file.
m.save("map.html")
```

```python
# Determine locations of users likely to use Elite Party Services
m2 = folium.Map(location=[32.7157, -117.1611], zoom_start=9)
users_df = pd.read_excel("1 - Elite Party Services - M7 - Geographical Data.xlsx", sheet_name="User") # Read user sheet from 1 - Elite Party Services - M7 - Geographical Data.xlsx file to get user id which is as user column 
M3 = pd.read_excel("1 - Elite Party Services - M3 - Data Composition.xlsx", sheet_name="1 - Elite Party Services - M3")# Read likelihood data from 1 - Elite Party Services - M3 - Data Composition.xlsx file to get user id which is as Q8 column 
coordinates['user'] = users_df['User']
coordinates
df = coordinates.merge(M3, left_on = 'user', right_on = 'ID', how = 'inner') # merging columns with id. resultant will have user, Latitude, Longitude
df
df_likelihood = df[df['Q8'] > 0] # Filtering only users likely to use Elite Party Services
df_likelihood

# Use a for loop to iterate over the rows of the DataFrame and add markers.
for index, row in df_likelihood.iterrows():
    latitude = row["Latitude"]
    longitude = row["Longitude"]
    folium.Marker(location=[latitude, longitude]).add_to(m2)

# Use the save() function to save the map to an HTML file.
m2.save("likely_map.html")
df_likelihood
```
